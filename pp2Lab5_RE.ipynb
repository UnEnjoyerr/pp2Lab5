```python
#1
import re
def find_matches_in_file(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        content = file.read()
    pattern = r'a(b*)'
    matches = re.findall(pattern, content)
    for match in matches:
        print(f"Found match: 'a{match}'")
filename = 'row.txt'
find_matches_in_file(filename)
```

```python
#4
import re
def find_sequences_in_file(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        content = file.read()
    pattern = r'[A-Z][a-z]+'
    matches = re.findall(pattern, content)
    for match in matches:
        print(f"Found match: '{match}'")
filename = 'row.txt'
find_sequences_in_file(filename)
```

```python
#6
import re
def replace_characters_in_file(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        content = file.read()
    pattern = r'[ ,.]'
    modified_content = re.sub(pattern, ':', content)
    print("Modified content:")
    print(modified_content)
filename = 'row.txt'
replace_characters_in_file(filename)
```

```python
#7
def snake_to_camel(snake_str):
    components = snake_str.split('_')
    camel_case_str = components[0] + ''.join(word.capitalize() for word in components[1:])
    return camel_case_str

lines = """this_is_a_test
another_example_string
convert_this_to_camel_case"""

for line in lines.splitlines():
    snake_case_string = line.strip() 
    if snake_case_string:
        camel_case_string = snake_to_camel(snake_case_string)
        print(f"Snake case: {snake_case_string} -> Camel case: {camel_case_string}")
```

```python
#8
import re
def split_at_uppercase(s):
    return re.split(r'(?=[A-Z])', s)
def split_strings_in_file(filename):
        with open(filename, 'r', encoding='utf-8') as file:
            lines = file.readlines()
        for line in lines:
            original_string = line.strip() 
            if original_string: 
                split_strings = split_at_uppercase(original_string)
                print(f"Original string: '{original_string}' -> Split: {split_strings}")

filename = 'row.txt'
split_strings_in_file(filename)
```

```python
#9
import re
def insert_spaces(s):
    return re.sub(r'(?<!^)(?=[A-Z])', ' ', s)
def insert_spaces_in_file(filename):
        with open(filename, 'r', encoding='utf-8') as file:
            lines = file.readlines()
        for line in lines:
            original_string = line.strip() 
            if original_string:  
                spaced_string = insert_spaces(original_string)
                print(f"Original string: '{original_string}' -> Spaced: '{spaced_string}'")
filename = 'row.txt'
insert_spaces_in_file(filename)
```

```python
#10
import re
def camel_to_snake(camel_str):
    snake_str = re.sub(r'(?<!^)(?=[A-Z])', '_', camel_str).lower()
    return snake_str
lines = """thisIsATest
anotherExampleString
convertThisToSnakeCase"""
for line in lines.splitlines():
    camel_case_string = line.strip() 
    if camel_case_string:
        snake_case_string = camel_to_snake(camel_case_string)
        print(f"Camel case: {camel_case_string} -> Snake case: {snake_case_string}")
```
